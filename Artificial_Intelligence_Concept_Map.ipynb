```python
import matplotlib.pyplot as plt
```

```python
import networkx as nx
```

```python
# Zihin haritasını oluşturmak için bir grafik yapısı kullanalım
G = nx.Graph()
```

```python
# Ana başlık: Yapay Zeka
G.add_node("Yapay Zeka (AI)")
```

```python

# Makine Öğrenimi ve diğer dalları
G.add_edges_from([
    ("Yapay Zeka (AI)", "Makine Öğrenimi (ML)"),
    ("Yapay Zeka (AI)", "Doğal Dil İşleme (NLP)"),
    ("Yapay Zeka (AI)", "Görüntü Tanıma (Computer Vision)"),

    # Makine Öğrenimi Dalları
    ("Makine Öğrenimi (ML)", "Yapay Sinir Ağları (ANN)"),
    ("Makine Öğrenimi (ML)", "Derin Öğrenme (Deep Learning)"),

    # Yapay Sinir Ağları detayları
    ("Yapay Sinir Ağları (ANN)", "Gizli Katmanlar (Hidden Layers)"),
    ("Yapay Sinir Ağları (ANN)", "Aktivasyon Fonksiyonları"),
    ("Yapay Sinir Ağları (ANN)", "Geri Yayılım (Backpropagation)"),

    # Derin Öğrenme detayları
    ("Derin Öğrenme (Deep Learning)", "Convolutional Neural Networks (CNN)"),
    ("Derin Öğrenme (Deep Learning)", "Recurrent Neural Networks (RNN)"),

    # NLP detayları
    ("Doğal Dil İşleme (NLP)", "Duygu Analizi (Sentiment Analysis)"),
    ("Doğal Dil İşleme (NLP)", "Dil Modellemesi (Language Modeling)"),

    # Görüntü Tanıma detayları
    ("Görüntü Tanıma (Computer Vision)", "Nesne Tanıma (Object Recognition)"),
    ("Görüntü Tanıma (Computer Vision)", "Yüz Tanıma (Face Recognition)"),
])
```

```python
# Mind map'i çizelim
pos = nx.spring_layout(G, seed=42)  # Konum düzenlemesi
plt.figure(figsize=(10, 8))
```

```python

# Kenarları ve düğümleri çiz
nx.draw(G, pos, with_labels=True, node_color="pink", node_size=3000, font_size=6, font_weight="bold", edge_color='gray')
```

```python
plt.title("Yapay Zeka ve İlgili Konu Başlıkları: Mind Map", size=14)
plt.show()
```
